# Data Meltdown

```python
import pandas as pd

# Sample Dataframe
data = pd.DataFrame({
    'Product' : ['A', 'B', 'C', 'D', 'E'],
    'Q1_Sales' : [100, 150, 200, 250, 300],
    'Q2_Sales' : [110, 160, 210, 260, 310],
    'Q3_Sales' : [120, 170, 220, 270, 320],
    'Q4_Sales' : [130, 180, 230, 280, 330],
})

print("Original Data (Wide Format):\n", data)

# Melt the dataframe to convertinto a long format
melted_data = data.melt(id_vars=['Product'],               # Identify the id variable
                        value_vars=['Q1_Sales', 'Q2_Sales', 'Q3_Sales', 'Q4_Sales'],  # Identify the value variables
                        var_name='Quarter',  # Name of the variable that will hold the quarter information
                        value_name='Sales')  # Name of the variable that will hold the sales figures

print("\nMelted Data (Long Format):\n")
display(melted_data)
```
