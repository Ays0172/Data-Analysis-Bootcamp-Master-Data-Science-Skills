```python
arr = np.array([5,6,7,8,9])
arr
```

```python
arr1 = np.append(arr, 7) # Append 7 to the array
arr1
```

```python
arr2 = np.insert(arr, 0, [1,2,3]) # Insert 1, 2, 3 at the beginning; index 0
arr2
```

```python
arr3 =np.delete(arr2,[]) # Delete elements at specified indices; here, no indices are specified, so no deletion occurs
arr3
```

```python
arr4 =np.delete(arr2,[2]) # Delete element at index 2
arr4
```

```python
arr5 = np.delete(arr4,[0,1]) # Delete elements at indices 0 and 1
arr5
```

```python
arrG = arr3.copy()  # Create a copy of arr3
arrG
```

Concatenating Arrays

```python
arr_concat = np.concatenate((arr1, arr2, arr3, arr4, arr5, arrG))
arr_concat
```

```python
a = np.array([[1,2,3,],[4,5,6]])
b = np.array([[6,5,4],[3,2,1]])

c = np.concatenate((a,b))
print(c)
```

```python
d = np.concatenate((a,b), axis = 0)  # Concatenate along the first axis (rows)
print(d)
```

```python
e = np.concatenate((a,b), axis = 1)  # Concatenate along the second axis (columns)
print(e)
```

V Stack

```python
x = np.vstack((a,b))  # Vertical stack
print(x)

y = np.hstack((a,b))  # Horizontal stack
print('\n', y)
```
