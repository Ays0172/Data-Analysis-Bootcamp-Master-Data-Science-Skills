```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
print(arr1)
```

To see the type of data

```python
print(type(arr1))
```

To see the shape (How many values within the variable)

```python
print(arr1.shape)
```

Check the value with in the variable at **index 4**

```python
print(arr1[4])  # Index starts from 0
```

To change the value at **index 3** to '9'

```python
arr1[3] = 9     # Update the 4th element
print(arr1)
```

If I want to change the whole array (2D Array)

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2)
```

To get specific values from 2D arrays

```python
arr2[0, 2]
```

```python
print(arr2.shape)
```

Fetching values using negetive indexing

```python
arr2[1][-2]
```

```python
arrC = (['India', 'Canada', 'USA', 'Australia', 'Germany'])
print(arrC)
print(type(arrC))
print(np.shape(arrC))
```

To create a list starting with '0' upto '29' with a increment of '3'

```python
arrM = np.arange(0, 30, 3)
print(arrM)
```

To generate an array of evenly spaced numbers

```python
arrL = np.linspace(0, 5, 10)  # 10 values between 0 and 5
print(arrL)
```

```python
arrL = np.linspace(0, 15, 30)  # 30 values between 0 and 15
print(arrL)
```

If you need an array with random values

```python
arr = np.random.rand(8)  # 1D array with 8 random values
print(arr)
```

```python
arr = np.random.rand(3, 4)  # 2D array with 3 rows and 4 columns
print(arr)
```

To cerate random values with 0 or 1

```python
print(np.zeros(8))
print('/y')
print(np.zeros((3,4)))
```

```python
print(np.ones((4,2,4))) # 3D array with 4 blocks, each having 2 rows and 4 columns
```

```python
arr = [0,1,2, 'India', 'A']
print(np.repeat(arr, 3))  # Repeat each element 3 times
```
