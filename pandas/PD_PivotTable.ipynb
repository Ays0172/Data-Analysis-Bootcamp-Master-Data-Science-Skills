```python
import pandas as pd

# Creating a Sample DataFrame
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09'],
    'Product': ['X', 'X', 'Y', 'Y', 'Z', 'Z', 'X', 'Y', 'Z'],
    'Region': ['North', 'North', 'South', 'South', 'East', 'East', 'West', 'West', 'West'],
    'Units': [10, 20, 30, 40, 50, 60, 70, 80, 90],
    'Sales': ['Jack', 'Sim', 'Tim', 'Sim', 'Jack', 'Sim', 'Tim', 'Sim', 'Jack'],
    'Profit': [100, 200, 300, 400, 500, 600, 700, 800, 900]
}

df = pd.DataFrame(data)

print("Original dataframe: ",)
display(df)

# Creating a Pivot Table
# Total Units by Product and Region
pivot_table_units = pd.pivot_table(df, index='Product', columns='Region', values='Units', aggfunc='sum', fill_value=0)
print("Pivot Table (Units): Total Units by Product and Region")
display(pivot_table_units)

# Total Profit by Product and Sales
pivot_table_profit = pd.pivot_table(df, index='Product', columns='Sales', values='Profit', aggfunc='mean', fill_value=0)
print("Pivot Table (Profit): Average Profit per Sales for each Product")
display(pivot_table_profit)

# Total and Average Units by Product for each Region
pivot_table_double = pd.pivot_table(df, index='Product', columns='Region', values='Units', aggfunc=['sum', 'mean'], fill_value=0)
print("Pivot Table (Units): Total and Average Units by Product for each Region")
display(pivot_table_double)
```
