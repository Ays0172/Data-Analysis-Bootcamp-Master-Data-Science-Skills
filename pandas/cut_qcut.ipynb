```python
import pandas as pd
import numpy as np

# Genereate a sample dataset
np.random.seed(0)
n = 1000
data = {
    'Customer': range(1, n+1),                              # range with 1000 customers
    'Age': np.random.randint(20, 60, n),                    # Random ages between 20 and 60
    'Annual_Spend': np.random.uniform(600, 20000, n),       # Random annual spend between 600 and 20000
    'Purchase_Frequency': np.random.poisson(2, n)           # Random purchase frequency with a Poisson distribution
}

df = pd.DataFrame(data)

# Step 1 - Create Age Groups
age_bins = [20, 30, 45, 60]  # Define age bins to match the age range in your data (20-60)
age_labels = ['Young', 'Middle-aged', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels) # df['Age_Group'] is now created

# Step 2 - Create Spending Segments
df['Spend_Segment'] = pd.qcut(df['Annual_Spend'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# Step 3 - Analyze Segments
segment_analysis = df.groupby(['Age_Group', 'Spend_Segment']).agg({
    'Customer': 'count',
    'Annual_Spend': 'mean',
    'Purchase_Frequency': ['mean', 'median', 'std']
}).reset_index()

# Renaming Columns
segment_analysis.columns = ['Age_Group', 'Spend_Segment', 'Cu'
'stomer_Count', 'Avg_Annual_Spend', 'Avg_Purchase_Frequency', 'Median_Purchase_Frequency', 'Std_Purchase_Frequency']

# Sort by Spend Segment and Age Group
segment_analysis = segment_analysis.sort_values(by=['Spend_Segment', 'Age_Group'])

print("Customer Segmentation Analysis:")
print(segment_analysis)
```
