# *Merge*

```python
import pandas as pd

df_id = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
})

df_orders = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Amount': [250, 150, 200, 300, 400],
    'Order ID' : [101, 102, 103, 104, 105]
})

# Merging the DataFrames on 'ID'
df_merged = pd.merge(df_id, df_orders, on='ID')
print("Merged DataFrame:")
display(df_merged)
```

# *Concat*

```python
df_sale_mar = pd.DataFrame({
    "Date" : ['2024-03-01', '2024-03-02', '2024-03-03', '2024-03-04', '2024-03-05'],
    "Sale" : [150, 200, 250, 300, 350]
})

df_sale_apr = pd.DataFrame({
    "Date" : ['2024-04-01', '2024-04-02', '2024-04-03', '2024-04-04', '2024-04-05'],
    "Sale" : [180, 220, 270, 320, 370]
})

# Concatenating the DataFrames
df_sales = pd.concat([df_sale_mar, df_sale_apr], ignore_index=True)  # Resetting index
print("Concatenated DataFrame:")
display(df_sales)
```

## *pd.merge and pd.concatenate*

```python
import pandas as pd

df_sale_mar = pd.DataFrame({
    'Product':['Apple', 'Banana', 'Orange', 'Mango', 'Grapes'],
    "Date" : ['2024-03-01', '2024-03-02', '2024-03-03', '2024-03-04', '2024-03-05'],
    "Sale" : [150, 200, 250, 300, 350]
})

df_sale_apr = pd.DataFrame({
    "Product':['Apple', 'Banana', 'Orange', 'Mango', 'Grapes'],"
    "Date" : ['2024-04-01', '2024-04-02', '2024-04-03', '2024-04-04', '2024-04-05'],
    "Sale" : [180, 220, 270, 320, 370]
})

df_id = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
})

df_orders = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Amount': [250, 150, 200, 300, 400],
    'Order ID' : [101, 102, 103, 104, 105]
})

# Step 1
df_sale = pd.concat([df_sale_mar, df_sale_apr], ignore_index=True)  # Resetting index
print("Combined DataFrame:")
print(df_sale)

# Step 2
df_final = pd.merge(df_sale, df_orders, on='Product', how='left')
print("Final Merged DataFrame:")
print(df_final)
```
