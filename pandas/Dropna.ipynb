```python
import pandas as pd
import numpy as np

# Creating Sample Datasets
a = pd.DataFrame({
    'A': [1, 2, 6, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [np.nan, 2, 3, 4, np.nan]
})

# Displaying the original DataFrame
print("DataFrame:\n", a)

# Dropping rows with any NaN values
df_dropna_rows = a.dropna()
print("\nDataFrame after dropping rows with any NaN values:")
display(df_dropna_rows)

# Dropping columns with any NaN values
df_dropna_columns = a.dropna(axis=1)
print("\nDataFrame after dropping columns with any NaN values:")
display(df_dropna_columns)
```

```python
df_dropna_specific = a.dropna(subset=['B', 'C'])
print("\nDataFrame after dropping rows with NaN values in specific columns (B and C):")
display(df_dropna_specific)
```

```python
# Dropping rows with NaN values in specific columns (B)
df_dropna_specific = a.dropna(subset=['B'])
print("\nDataFrame after dropping rows with NaN values in specific columns (B):")
display(df_dropna_specific)
```
