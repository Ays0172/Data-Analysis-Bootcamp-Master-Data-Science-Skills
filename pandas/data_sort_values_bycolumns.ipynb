```python
# Importing Libraries
import pandas as pd
import numpy as np

# Extend sample dataframe
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Age': np.random.randint(20, 40, size= 10),
    'Score': np.random.randint(35, 100, size= 10),
    'Country': np.random.choice(['USA', 'Canada', 'UK', 'Germany', 'France'], size=10),
    'Salary': np.random.randint(40000, 120000, size=10)
})

print("Original DataFrame:")
print(data)
```

```python
# Sorting data by Country
sorted_data = data.sort_values(by=['Country', 'Score'], ascending=[True, False]) 
print("\nSorted DataFrame by Country and Score (Score in descending order and Country in Ascending Order): ")
display(sorted_data)
```

```python
# Sorting data by Age
sorted_data = data.sort_values(by=['Age', 'Salary'], ascending=[True, False])
print("\nSorted DataFrame by Age (Age in Ascending Order and Salary in Descending Order): ")
display(sorted_data)
```
