```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# generating Sample Data
date_range = pd.date_range(start = "2024-01-01", end = "2024-06-30", freq = "H")
np.random.seed(0)
energy_data = np.random.randint(100, 500, size=len(date_range))

# Creating DataFrame
df = pd.DataFrame({
 "Timestamp": date_range,
 "Energy_Consumption_KWh": energy_data
})
df.set_index('Timestamp', inplace = True)

# Step 1
daily_consumption = df.resample('D').sum()
daily_consumption.columns = ['Daily_Total_KWh']

# Step 2
weekly_consumption = daily_consumption.resample('W').mean()
weekly_consumption.columns = ['Weekly_Average_KWh']

# Visualisation


# Daily Total
plt.figure(figsize=(15, 5))
plt.plot(daily_consumption.index, daily_consumption['Daily_Total_KWh'], marker='o', linestyle='-', color = 'red', label = 'Daily Total KWh')
plt.title('Daily Energy Consumption (KWh)')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (KWh)')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Weekly Average
plt.figure(figsize=(12, 6))
plt.plot(weekly_consumption.index, weekly_consumption['Weekly_Average_KWh'], marker='*', linestyle='-')
plt.title('Weekly Average Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (KWh)')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
