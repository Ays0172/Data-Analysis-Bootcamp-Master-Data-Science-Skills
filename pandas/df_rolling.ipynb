```python
import pandas as pd
import numpy as np

# Generate a Sample dataset [Timeseries]
np.random.seed(45)
dates = pd.date_range(start='2024-01-01', periods=1100)
data = {'Date' : dates,
        'Sale' : np.random.randint(99, 198, size = 1100)}

df = pd.DataFrame(data)
df.set_index('Date', inplace = True)

# Calculating 7 day average sale
df['7-day MA'] = df['Sale'].rolling(window=7).mean().round(2)   # 2 Decimal places

# calculating 7 Day rolling standard deviation
df['7-day std'] = df['Sale'].rolling(window=7).std().round(2)   # 2 Decimal places

# Display the DataFrame
print("Sample Data ")
display(df.head(14))
```
